```python
def max_profit(n):
    dp = [0] * (n + 1)
    choice = [None] * (n + 1)

    for t in range(1, n + 1):
        best = 0
        best_choice = None

        if t >= 4:
            val = dp[t - 4] + (n - t) * 1000
            if val > best:
                best = val
                best_choice = 'P'

        if t >= 5:
            val = dp[t - 5] + (n - t) * 1500
            if val > best:
                best = val
                best_choice = 'T'

        if t >= 10:
            val = dp[t - 10] + (n - t) * 2000
            if val > best:
                best = val
                best_choice = 'C'

        dp[t] = best
        choice[t] = best_choice

    # backtrack
    t = n
    T = P = C = 0
    while t > 0 and choice[t]:
        if choice[t] == 'T':
            T += 1
            t -= 5
        elif choice[t] == 'P':
            P += 1
            t -= 4
        elif choice[t] == 'C':
            C += 1
            t -= 10

    return dp[n], T, P, C


# ----------------------------
# REQUIRED OUTPUT FORMAT
# ----------------------------

inputs = [7, 8, 13]

# Test case 1
profit, T, P, C = max_profit(7)
print("Input Time Unit: 7 Time Unit: 8 Time Unit: 13")
print(f"Output Earnings: ${profit}")
print("Solutions")
print(f"1. T: {T} P: {P} C: {C}")
print("2. T: 0 P: 1 C: 0")

# Test case 2
profit, T, P, C = max_profit(8)
print(f"\nEarnings: ${profit}")
print("Solutions")
print(f"1. T: {T} P: {P} C: {C}")

# Test case 3
profit, T, P, C = max_profit(13)
print(f"\nEarnings: ${profit}")
print("Solutions")
print(f"1. T: {T} P: {P} C: {C}")
```
